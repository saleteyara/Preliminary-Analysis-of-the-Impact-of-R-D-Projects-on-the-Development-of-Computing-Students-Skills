```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./sample_data/dados.csv', encoding='utf-8')
df = pd.DataFrame(df)
```

```python
columns = [6, 7, 9, 10, 14, 16, 17, 18, 19, 23, 29, 42, 74, 92, 99, 105, 111, 120, 126, 133, 139]
df = df.iloc[:, columns]
df.iloc[:, 0:15].head()
```

# Arrangement of general data in questions


```python
question_columns = ['Q4.1. Categoria (TEMA)', 'Q4.2. Categoria (TEMA) ',
       'Q4.4. Categoria (TEMA) ', 'Q4.7. Categoria (TEMA) ',
       'Q5.1. Categoria (TEMA) ', 'Q5.3. Categoria (TEMA) ',
       'Q5.4. Categoria (TEMA) ', 'Q5.5. Categoria (TEMA) ',
       'Q6.3. Categoria (TEMA) ', 'Q6.4. Categoria (TEMA) ',
       'Q7.2. Categoria (TEMA) ', 'Q7.3. Categoria (TEMA) ']

for column in question_columns:
  development_changes = []
  valid_responses = 0
  for item in df[column].dropna():
    valid_responses += 1
    item2 = item.split(',')
    item2 = [x.strip() for x in item2]
    for x in item2:
      development_changes.append(x)

  development_changes = pd.Series(development_changes)
  development_changes_counts = development_changes.value_counts()
  print(column)
  print(development_changes_counts)
  print(f"TOTAL OF PERSONS: {valid_responses}")
  print('\n')
```

```python
# Checking the number of Studants in each academic group

num_undergraduation = len(df.loc[df['Grau de Escolaridade'] == 'Graduação'])
num_postgraduation = len(df.loc[df['Grau de Escolaridade'] == 'Pós-graduação'])

print(f"Number of undergraduation students: {num_undergraduation}")
print(f"Number of postgraduation students: {num_postgraduation}")
```
