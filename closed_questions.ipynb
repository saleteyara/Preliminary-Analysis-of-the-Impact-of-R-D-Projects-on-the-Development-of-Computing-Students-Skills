```python
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def processar(df, nome_grupo):
    Q48 = df[df.filter(regex='^Q4.8.').columns]
    results = {}
    for col in Q48:
        match = re.search(r'\[(.*?)\]', col)
        if match:
            categoria = match.group(1)
            contagem = df[col].value_counts().sort_index()
            results[categoria] = contagem
    resultado = pd.DataFrame(results).fillna(0).astype(int).T
    resultado["Grupo"] = nome_grupo
    return resultado

# ======================
# Construindo resultados
# ======================
resultado1 = processar(df_tempo_projeto1, "Menos de 1 ano")
resultado2 = processar(df_tempo_projeto2, "1 a 2 anos")
resultado3 = processar(df_tempo_projeto3, "3 a 4 anos")
resultado4 = processar(df_tempo_projeto4, "5 anos ou mais")

# ======================
# Juntando todos
# ======================
comparacao = pd.concat([resultado1, resultado2, resultado3, resultado4])
comparacao = comparacao.reset_index().rename(columns={"index": "Competência"})

# Transformar em formato longo
melted = comparacao.melt(
    id_vars=["Grupo", "Competência"],
    var_name="Nivel",
    value_name="Frequencia"
)
melted["Nivel"] = melted["Nivel"].astype(int)

# ======================
# Média ponderada por grupo/competência
# ======================
media = (melted.groupby(["Grupo", "Competência"])
         .apply(lambda x: (x["Nivel"] * x["Frequencia"]).sum() / x["Frequencia"].sum()
                if x["Frequencia"].sum() > 0 else 0)
         .reset_index(name="Média"))

# ======================
# Pivot para heatmap
# ======================
pivot = media.pivot(index="Grupo", columns="Competência", values="Média")

# ======================
# Plotando
# ======================
plt.figure(figsize=(14,6))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", cbar=True, fmt=".1f")

plt.title("Média de Desenvolvimento de Competências por Tempo de Projeto")
plt.xlabel("Competências")
plt.ylabel("Grupo")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
